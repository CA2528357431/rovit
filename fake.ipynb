```python
import matplotlib.pyplot as plt

# 数据
x = [10, 30, 50, 70, 100]  # x轴数据
y1 = [67.68, 68.45, 69.02, 69.60, 70.12]  # 第二条折线数据
y2 = [70.81, 71.47, 71.88, 72.24, 72.39]  # 第一条折线数据

# 创建图表
plt.figure(figsize=(8, 6))

# 绘制第一条折线
plt.plot(x, y1, label='ViT', marker='o')

# 绘制第二条折线
plt.plot(x, y2, label='RoViT', marker='s')

# 添加标题和标签
plt.title('Accuracy VS Label Amount')
plt.xlabel('Label Amount')
plt.ylabel('Accuracy')
plt.legend()


# 显示图表
plt.tight_layout()
plt.show()
```
